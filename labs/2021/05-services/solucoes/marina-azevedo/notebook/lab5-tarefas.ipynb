## Tarefa 1

### Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
import io
import requests
import json
```

```python
urlTemplate = "http://dbpedia.org/data/{concept}.{format}"

concept = "Tent of Miracles (novel)"
format = "jsod"

concept = concept.replace(" ", "_")

url = urlTemplate.replace("{concept}", concept)\
                  .replace("{format}", format)

data = requests.get(url)
js = data.json()
print(json.dumps(js, indent=3))
```

## Tarefa 2
### Recupere via Python e mostre o autor de Tenda dos Milagres?

```python
author = js["d"]["results"][0]["http://dbpedia.org/property/author"]["__deferred"]["uri"]
print("Autor de Tenda dos Milagres: " + author)
```

## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
urlTemplate = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label={concept}&match={match}&limit={limit}"

concept = "Aspirin"
match = "exact"
limit = "10"

concept = concept.replace(" ", "%20")

url = urlTemplate.replace("{concept}", concept)\
                  .replace("{match}", match)\
                  .replace("{limit}", limit)

data = requests.get(url)
jsData = data.json()
print(json.dumps(jsData, indent=3))
```

```python
uri = jsData[0]["resource"]
print("uri: " + uri)
bar = str.rindex(uri, "/")
print("código: " + uri[bar+1:])
```

```python
urlTemplate = "https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code}"

code = uri[bar+1:]

url = urlTemplate.replace("{code}", code)

data = requests.get(url)
jsData = data.json()
print(json.dumps(jsData, indent=3))
```

## Tarefa 4
### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

```python
urlTemplate = "http://nominatim.openstreetmap.org/search?format={format}&q={query}"
format = "json"
query = "Country Music Hall of Fame and Museum"

url = urlTemplate.replace("{format}", format)\
                  .replace("{query}", query)

data = requests.get(url)
museum_ref = data.json()
print(json.dumps(museum_ref, indent=3))
```

```python
print(museum_ref[0]["osm_id"])
```

```python
urlTemplate = "https://api.openstreetmap.org/api/0.6/{element_type}/{id}.{format}"

element_type = "way" # types: node, way, or relation
id = str(museum_ref[0]["osm_id"])
format = "json"

url = urlTemplate.replace("{element_type}", element_type)\
                  .replace("{id}", id)\
                  .replace("{format}", format)

data = requests.get(url)
museum_way = data.json()
print(json.dumps(museum_way, indent=3))
```

```python
print(museum_way["elements"][0]["tags"]["wikipedia"])
```
