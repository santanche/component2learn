## Tarefa 1

### Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
import requests
import json

# Lucas Franzolin - 150621
```

```python
concept = "Tent of Miracles (novel)"
req_format = "jsod"
url = "http://dbpedia.org/data/{}.{}".format(concept.replace(" ", "_"), req_format)
data = requests.get(url)
gson0 = data.json()
print(json.dumps(gson0, indent=2, default=str))
```

## Tarefa 2
### Recupere via Python e mostre o autor de Tenda dos Milagres?

```python
author = gson0["d"]["results"][0]["http://dbpedia.org/property/author"]["__deferred"]["uri"]
print("Autor: {}".format(author))
```

## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
concept = "Aspirin"
match = "exact"
limit = "10"
url = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label={}&match={}&limit={}".format(concept.replace(" ", "%20"), match, limit)
data = requests.get(url)
gson1 = data.json()
print(json.dumps(gson1, indent=2, default=str))
```

```python
uri = gson1[0]["resource"]
bar = str.rindex(uri, "/")
print("uri: {}\ncode: {}".format(uri, uri[bar+1:]))
```

```python
code = uri[bar+1:]
url = "https://id.nlm.nih.gov/mesh/lookup/details?descriptor={}".format(code)
data = requests.get(url)
gson2 = data.json()
print(json.dumps(gson2, indent=2, default=str))
```

## Tarefa 4
### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

```python
req_format = "json"
query = "Country Music Hall of Fame and Museum"
url = "http://nominatim.openstreetmap.org/search?format={}&q={}".format(req_format, query)
data = requests.get(url)
gson3 = data.json()
print(json.dumps(gson3, indent=2, default=str))
```

```python
gson3[0]["osm_id"]
```

```python
element_type = "way"
id = str(gson3[0]["osm_id"])
url = "https://api.openstreetmap.org/api/0.6/{}/{}.{}".format(element_type, id, req_format)

data = requests.get(url)
gson4 = data.json()
print(json.dumps(gson4, indent=2, default=str))
```
