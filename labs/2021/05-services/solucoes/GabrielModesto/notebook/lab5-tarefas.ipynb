## Tarefa 1


### Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
import io
import requests
import json
```

```python
url_template = "http://dbpedia.org/data/{novelName}.jsod"

novelName = "Tent of Miracles (novel)"

novelName = novelName.replace(" ","_")

url_drum = url_template.replace("{novelName}", novelName)

data_drum = requests.get(url_drum)
json_drum = data_drum.json()

print(json.dumps(json_drum, indent=3))
```

## Tarefa 2
### Recupere via Python e mostre o autor de Tenda dos Milagres?

```python
grass = json_drum["d"]["results"][0]["http://dbpedia.org/ontology/author"]["__deferred"]["uri"]
print("Autor de Tenda dos Milagres: " + grass)
```

## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label={label}&match={match}&limit={limit}"

label = "Aspirin"
match = "exact"
limit = "10"

label = label.replace(" ","%20")

url = url_template.replace("{label}",label)\
                  .replace("{match}",match)\
                  .replace("{limit}",limit)

data_drum = requests.get(url)
json_drum = data_drum.json()
print(json.dumps(json_drum, indent=3))
```

```python
uri_mesh = json_drum[0]["resource"]
print("uri do recurso: " + uri_mesh)
code = str.rindex(uri_mesh, "/")
print("código do recurso: " + uri_mesh[code+1:])
```

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code}"

code = uri_mesh[code+1:]

url = url_template.replace("{code}",code)

data_drum = requests.get(url)
json_drum = data_drum.json()
print(json.dumps(json_drum, indent=3))
```

### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

```python
url_template = "http://nominatim.openstreetmap.org/search?format={format}&q={query}"
format = "json"
query = "Museu do Ipiranga"

url = url_template.replace("{format}", format)\
                  .replace("{query}", query)

data_drum = requests.get(url)
json_drum = data_drum.json()
print(json.dumps(json_drum, indent=3))
```

```python
print(json_drum[0]["osm_id"])
```

```python
url_template = "https://api.openstreetmap.org/api/0.6/{element_type}/{id}.{format}"

element_type = "way" 
id = str(json_drum[0]["osm_id"])
format = "json"

url = url_template.replace("{element_type}", element_type)\
                  .replace("{id}", id)\
                  .replace("{format}", format)

data_drum = requests.get(url)
place_way = data_drum.json()
print(json.dumps(place_way, indent=3))
```

```python
print(place_way["elements"][0]["tags"]["wikipedia"])
```

```python
url_template = "http://dbpedia.org/data/{concept}.{format}"
concept = place_way["elements"][0]["tags"]["wikipedia"].replace(" ", "_")\
                                                        .replace("pt:", "")
format = "jsod"

url = url_template.replace("{concept}", concept)\
                  .replace("{format}", format)

data = requests.get(url)
place_dbpedia = data.json()
print(json.dumps(place_dbpedia, indent=3))
```
