## Tarefa 1

### Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
import requests
import json
```

```python
url_template = "http://dbpedia.org/data/{concept}.{format}"

concept = "Tent of Miracles (novel)"
format = "jsod"

concept = concept.replace(" ", "_")

url = url_template.replace("{concept}", concept)\
                  .replace("{format}", format)

data = requests.get(url)
tent_of_miracles_json = data.json()

print(json.dumps(tent_of_miracles_json, indent=3))
```

## Tarefa 2
### Recupere via Python e mostre o autor de Tenda dos Milagres?

```python
# fetch author URI
author_uri = tent_of_miracles_json["d"]["results"][0]["http://dbpedia.org/property/author"]["__deferred"]["uri"]
print('URL do autor: ' + author_uri)

# process URL
url_template = "http://dbpedia.org/data/{concept}.{format}"

concept = author_uri.split("/")[-1]
format = "jsod"

concept = concept.replace(" ", "_")

url = url_template.replace("{concept}", concept)\
                  .replace("{format}", format)

# fetch author information
author_data = requests.get(url)
author_json = author_data.json()

# process author name
author_name = author_json["d"]["results"][0]["http://dbpedia.org/property/name"]

print('Autor de Tenda dos Milagres: ' + author_name)
```

## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label={concept}&match={match}&limit={limit}"

concept = "Aspirin"
match = "exact"
limit = "10"

concept = concept.replace(" ", "%20")

url = url_template.replace("{concept}", concept)\
                  .replace("{match}", match)\
                  .replace("{limit}", limit)

data = requests.get(url)
json_mi = data.json()
print(json.dumps(json_mi, indent=3))
```

```python
uri_mesh = json_mi[0]["resource"]
print("uri do recurso: " + uri_mesh)
bar = str.rindex(uri_mesh, "/")
print("código do recurso: " + uri_mesh[bar+1:])
```

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code}"

code = uri_mesh[bar+1:]

url = url_template.replace("{code}", code)

data = requests.get(url)
json_mid = data.json()
print(json.dumps(json_mid, indent=3))
```

## Tarefa 4
### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

```python
url_template = "http://nominatim.openstreetmap.org/search?format={format}&q={query}"
format = "json"
query = "Empire State Building"

url = url_template.replace("{format}", format)\
                  .replace("{query}", query)

# fetch data from OpenStreetMap API
data = requests.get(url)
building_ref = data.json()
print(json.dumps(building_ref, indent=3))
```

```python
# display OpenStreetMap id of Empire State Building
print(building_ref[0]["osm_id"])

# build URL
url_template = "https://api.openstreetmap.org/api/0.6/{element_type}/{id}.{format}"

element_type = "way" # types: node, way, or relation
id = str(building_ref[0]["osm_id"])
format = "json"

url = url_template.replace("{element_type}", element_type)\
                  .replace("{id}", id)\
                  .replace("{format}", format)

# fetch elements from OpenStreetMap API for Empire State Building
data = requests.get(url)
building_way = data.json()
print(json.dumps(building_way, indent=3))
```

```python
# display Wikipedia tag
print(building_way["elements"][0]["tags"]["wikipedia"])

# build URL
url_template = "http://dbpedia.org/data/{concept}.{format}"
concept = building_way["elements"][0]["tags"]["wikipedia"].replace(" ", "_")\
                                                        .replace("en:", "")
format = "jsod"

url = url_template.replace("{concept}", concept)\
                  .replace("{format}", format)

# fetch Empire State Building data from DBPedia
data = requests.get(url)
building_dbpedia = data.json()
print(json.dumps(building_dbpedia, indent=3))
```
