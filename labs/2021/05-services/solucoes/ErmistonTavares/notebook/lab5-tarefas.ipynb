## Tarefa 1

### Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
url_template = "http://dbpedia.org/data/{concept}.{format}"

concept = "Tent of Miracles (novel)"
format = "jsod"

concept = concept.replace(" ", "_")

url = url_template.replace("{concept}", concept)\
                  .replace("{format}", format)

data = requests.get(url)
js = data.json()
print(json.dumps(js, indent=3))
```

## Tarefa 2
### Recupere via Python e mostre o autor de Tenda dos Milagres?

```python
grass = js["d"]["results"][0]["http://dbpedia.org/ontology/author"]["__deferred"]["uri"]
print("Autor de Tent of Miracles: " + grass)
```

## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

## Buscando o código de Aspirina

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label={concept}&match={match}&limit={limit}"

concept = "Aspirin"
match = "exact"
limit = "10"

concept = concept.replace(" ", "%20")

url = url_template.replace("{concept}", concept)\
                  .replace("{match}", match)\
                  .replace("{limit}", limit)

data = requests.get(url)
json_mi = data.json()
print(json.dumps(json_mi, indent=3))
```

## Acessando o código do recurso

```python
uri_mesh = json_mi[0]["resource"]
print("uri do recurso: " + uri_mesh)
bar = str.rindex(uri_mesh, "/")
print("código do recurso: " + uri_mesh[bar+1:])
```

## Acessando os detalhes a partir do código

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code}"

code = uri_mesh[bar+1:]

url = url_template.replace("{code}", code)

data = requests.get(url)
json_mid = data.json()
print(json.dumps(json_mid, indent=3))
```

## Extraindo os termos para aspirina

```python
print("Termos usados para Aspirina: \n")
for terms in json_mid['terms']:
    print (terms["label"])
```

### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

## REST API do OpenStreetMap para Query

```python
url_template = "http://nominatim.openstreetmap.org/search?format={format}&q={query}"
format = "json"
query = "Gillette stadium"

url = url_template.replace("{format}", format)\
                  .replace("{query}", query)

data = requests.get(url)
stadium_ref = data.json()
print(json.dumps(stadium_ref, indent=3))
```

## Extraindo o Identificador OSM

```python
print(stadium_ref[0]["osm_id"])
```

## REST API do OpenStreetMap para Elementos (Node, Way e Relation)

```python
url_template = "https://api.openstreetmap.org/api/0.6/{element_type}/{id}.{format}"

element_type = "way" # types: node, way, or relation
id = str(stadium_ref[0]["osm_id"])
format = "json"

url = url_template.replace("{element_type}", element_type)\
                  .replace("{id}", id)\
                  .replace("{format}", format)

data = requests.get(url)
stadium_way = data.json()
print(json.dumps(stadium_way, indent=3))
```

## Extraindo o Identificador DBPedia

```python
print(stadium_way["elements"][0]["tags"]["wikipedia"])
```

## REST API do DBPedia

```python
url_template = "http://dbpedia.org/data/{concept}.{format}"
concept = stadium_way["elements"][0]["tags"]["wikipedia"].replace(" ", "_")\
                                                        .replace("en:", "")
format = "jsod"

url = url_template.replace("{concept}", concept)\
                  .replace("{format}", format)

data = requests.get(url)
stadium_dbpedia = data.json()
print(json.dumps(stadium_dbpedia, indent=3))
```
