## Tarefa 1

### Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
import io
import requests
import json
```

```python
url = "http://dbpedia.org/data/Tent_of_Miracles_(novel).jsod"
data = requests.get(url)
js = data.json()
print(json.dumps(js, indent = 3))
```

```python
url_template = "http://dbpedia.org/data/{node}.{type}"

node = "Tent of Miracles (novel)"
type = "jsod"

node = node.replace(" ", "_")

url = url_template.replace("{node}", node)\
                  .replace("{type}", type)

data = requests.get(url)
js = data.json()
print(json.dumps(js, indent = 3))
```

```python
reference = js["d"]["results"][0]["http://dbpedia.org/ontology/author"]["__deferred"]["uri"]
print("The author of Tent of Miracles (novel) is: " + reference)
```

## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label={concept}&match={match}&limit={limit}"

concept = "Aspirin"
match = "exact"
limit = "10"

concept = concept.replace(" ", "%20")

url = url_template.replace("{concept}", concept)\
                  .replace("{match}", match)\
                  .replace("{limit}", limit)

data = requests.get(url)
json_mi = data.json()
print(json.dumps(json_mi, indent=3))
```

```python
uri_mesh = json_mi[0]["resource"]
print("Using uri: " + uri_mesh)
bar = str.rindex(uri_mesh, "/")
print("Resource code: " + uri_mesh[bar+1:])
```

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code}"

code = uri_mesh[bar+1:]

url = url_template.replace("{code}", code)

data = requests.get(url)
json_mid = data.json()
print(json.dumps(json_mid, indent=3))
```

```python
jsonObject = json.loads(json.dumps(json_mid, indent=3))

for doc in jsonObject['terms']:
   print(doc['label'])
            
    
```

## Tarefa 4
### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

```python
url_template = "http://nominatim.openstreetmap.org/search?format={format}&q={query}"
format = "json"
query = "Walt Disney World Resort, Bay Lake, Orange County, Florida, 32830, United States"

url = url_template.replace("{format}", format)\
                  .replace("{query}", query)

data = requests.get(url)
disney_ref = data.json()
print(json.dumps(disney_ref, indent=3))
```

```python
print(disney_ref[0]["osm_id"])
```

```python
url_template = "https://api.openstreetmap.org/api/0.6/{element_type}/{id}.{format}"

element_type = "relation" # types: node, way, or relation
id = str(disney_ref[0]["osm_id"])
format = "json"

url = url_template.replace("{element_type}", element_type)\
                  .replace("{id}", id)\
                  .replace("{format}", format)

data = requests.get(url)
disney_relation = data.json()
print(json.dumps(disney_relation, indent=3))
```

```python
print(disney_relation["elements"][0]["tags"]["wikipedia"])
```

```python
url_template = "http://dbpedia.org/data/{concept}.{format}"
concept = disney_relation["elements"][0]["tags"]["wikipedia"].replace(" ", "_")\
                                                        .replace("en:", "")
format = "jsod"

url = url_template.replace("{concept}", concept)\
                  .replace("{format}", format)

data = requests.get(url)
disney_dbpedia = data.json()
print(json.dumps(disney_dbpedia, indent=3))
```
