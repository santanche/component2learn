This notebook reads the `breast-cancer-wisconsin.csv` file and draws a scatter plot of `radius_mean` vs `fractal_dimension_mean`, coloring points by the `diagnosis` column (`B` or `M`).

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('breast-cancer-wisconsin.csv')

# Map diagnosis to colors
color_map = {'B': 'blue', 'M': 'red'}
colors = df['diagnosis'].map(color_map)

# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(df['radius_mean'], df['fractal_dimension_mean'], c=colors, alpha=0.7)
plt.xlabel('radius_mean')
plt.ylabel('fractal_dimension_mean')
plt.title('Breast Cancer: radius_mean vs fractal_dimension_mean')
plt.show()
```
