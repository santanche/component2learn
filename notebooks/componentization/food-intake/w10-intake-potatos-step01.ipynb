```python
import matplotlib.pyplot as plt

def scatter_plot(data, xfield, yfield, title):
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(data[xfield], data[yfield], c=data[yfield], cmap='viridis')
    
    plt.xlabel(xfield)
    plt.ylabel(yfield)
    plt.title(title)
    
    plt.colorbar(scatter, label=yfield)
    
    plt.show()
```
