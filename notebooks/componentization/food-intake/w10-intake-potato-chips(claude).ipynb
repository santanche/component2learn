```python
# Read a CSV file in the relative path ../../../data/food-intake/grouped/intake-person-demo(potato-chips).csv
# and print the first five rows of the file

import pandas as pd

# Read the CSV file
df = pd.read_csv('intake-person-demo(potato-chips).csv')

# Print the first five rows of the file
print(df.head())
```

```python
# Draw a scatter plot of this file with age vs. intake_bw
# changing the bullet color of intake_bw according to its value

import matplotlib.pyplot as plt

# Create a scatter plot
plt.scatter(df['age'], df['intake_bw'], c=df['intake_bw'], cmap='viridis')

# Add labels to the plot
plt.xlabel('Age')
plt.ylabel('Intake BW')
plt.title('Age vs. Intake BW')

# Show the plot
plt.show()
```
